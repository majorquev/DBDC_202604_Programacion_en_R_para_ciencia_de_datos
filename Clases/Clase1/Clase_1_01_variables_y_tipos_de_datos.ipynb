### 2.1. Missing values

En R, existen tipos de datos especiales para caracterizar la ausencia o la no definición de una variable. Los valores especiales que maneja R son NA, NaN, Inf, -Inf.

```
# 2.1. Missing Values ------------
# Acá se explica la diferencia entre ellos y cómo reconocerlos.
# NA: "Not Available" (SQL: NULL, Excel/PBI: Blank)
# NaN: "Not a Number", en particular es una NA, pero que además no es numérico. (0/0)
```

```
RUT <- c("1-x", NA, "3-k")
RUT
```

```
# Nos interesa poder identificar los casos NA, por ejemplo, en una query sobre una tabla con un campo llamado RUT
# RUT es NA
RUT == NA
RUT == "1-x"
```

```
3 == NA
NA == NA
```

```
# para identificar un valor NA o NaN podemos utilizar la función is.na
is.na(3)
is.na(NA)
is.na(RUT)
```

```
# Ojo. Un NaN es un NA, pero un NA no necesariamente es un NaN.

is.na(NaN)# SQL: IS NULL
is.nan(NaN)
is.nan(NA)
```

```
!is.na(NaN) # SQL: IS NOT NULL
```
